```python
#Data Encoding
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv("abalone.csv")
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data_encoded=df.apply(LabelEncoder().fit_transform)
print(data_encoded)
```

```python
#HANDLE MISSING VALUES
df.isnull().sum()
```

```python
#Spilt data into independent variables.
X = data_encoded.iloc[:, :-1].values
print(X)
```

```python
#Split into dependent variables
Y = data_encoded.iloc[:, -1].values
print(Y)
```

```python
#Scaling independent variables
from sklearn.preprocessing import StandardScaler
 
object= StandardScaler()
 
# standardization 
scale = object.fit_transform(X) 
print(scale)
```

```python
#FIND THE OUTLIERS
sns.boxplot(x='Diameter', data=df)
```

```python
#REMOVING THE OULIERS BASED ON SCORE
import numpy as np
import sklearn
from sklearn.datasets import load_boston 

Q1 = np.percentile(df['Diameter'], 25, interpolation = 'midpoint') 
Q3 = np.percentile(df['Diameter'], 75, interpolation = 'midpoint') 
IQR = Q3 - Q1 
print("Old Shape: ", df.shape) 
upper = np.where(df['Diameter'] >= (Q3+1.5*IQR)) 
lower = np.where(df['Diameter'] <= (Q1-1.5*IQR)) 
df.drop(upper[0], inplace = True) 
df.drop(lower[0], inplace = True) 
print("New Shape: ", df.shape) 
sns.boxplot(x='Diameter', data=df)
```
